البته بعضی فعالیت های شبیه به هم کمی با هم قاطی شدن. مثلا نشستن و ایستادن نزدیک هم افتادن چون از نظر حرکت بدن خیلی تفاوت زیادی ندارن. همینطور انواع راه رفتن هم کمی به هم نزدیکن چون همشون الگوی حرکتی مشابه دارن

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(label_train)
print("\nDataFrame after Label Encoding:")
print(train['Activity_Name'])
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(feature_train,y_encoded,test_size=0.2,random_state=42,stratify=y_encoded)
```

اکنون می‌توانید با استفاده از الگوریتم‌های یادگیری ماشین، مدلی آموزش دهید که با گرفتن مقادیر ویژگی‌های دریافتی از سنسورها، نوع فعالیت در حال انجام را پیش‌بینی کند.
در این قسمت از شما می‌خواهیم از الگوریتم درخت تصمیم یا الگوریتم های تجمیعی  استفاده کنید و ترکیب‌های مختلفی از هایپرپارامترهای مهم این الگوریتم هارا آزمایش کنید تا بهترین مدل را برای پیش‌بینی فعالیت‌های انسانی انتخاب کنید. در این قسمت می‌توانید از کتابخانه‌ی <code>sklearn</code> استفاده کنید و برای جست‌وجوی هایپرپارامترها می‌توانید از روشی همچون Grid Search استفاده کنید. منطقی است که برای پیدا کردن بهترین ترکیب هایپرپارامتر نیاز به مجموعه‌ی اعتبارسنجی یا استفاده از روش‌هایی مانند Cross Validation دارید.
</font>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python

parameters = {'n_estimators': range(100, 301, 50),'max_depth': range(5, 21, 5)}
rfc = RandomForestClassifier(random_state=42)
grid = GridSearchCV(estimator=rfc,param_grid=parameters,cv=5,scoring='f1_weighted',n_jobs=-1)
grid.fit(X_tr, y_tr)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score
```

```python
best_model=grid.best_estimator_
y_val_pred=best_model.predict(X_val)
print('Accuracy:',accuracy_score(y_val,y_val_pred))
print('F1-score (weighted):',f1_score(y_val,y_val_pred,average='weighted'))
print('ClassificationReport:')
print(classification_report(y_val,y_val_pred))
print('ConfusionMatrix:')
print(confusion_matrix(y_val,y_val_pred))
```
